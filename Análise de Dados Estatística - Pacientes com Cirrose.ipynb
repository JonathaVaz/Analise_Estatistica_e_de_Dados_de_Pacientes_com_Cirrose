### **Descrição dos Atributos**

##### ***ID*** - Identificador Exclusivo (Unico)

##### ***N_Days*** - Número de dias entre o registro e o que ocorrer antes do óbito, transplante ou análise do estudo em julho de 1986

##### ***Status*** - Status do Paciente [C (censurado), CL (censurado devido ao tx do fígado), D (óbito)]

##### ***Medicamento*** - Tipo de Medicamento D-penicilamina ou placebo

##### ***Idade*** - idade (em dias)

##### ***Sexo*** - M (masculino), F (feminino)

##### ***Ascite*** - presença de ascite N (Não), Y (Sim)

##### ***Hepatomegalia*** - presença de hepatomegalia N (Não), Y (Sim)

##### ***Aranhas*** – presença de aranhas N (Não), Y (Sim)

##### ***Edema*** - presença de edema N (sem edema e sem terapia diurética para edema), S (edema presente sem diuréticos ou edema resolvido com diuréticos), Y (edema apesar da terapia diurética)

##### ***Bilirrubina*** - bilirrubina sérica (mg/dL)

##### ***Colesterol*** - colesterol sérico (mg/dL)

##### ***Albumina*** - albumina (gm/dL)

##### ***Cobre*** - cobre na urina (ug/dia)

##### ***Alk_Phos*** - fosfatase alcalina (U/litro)

##### ***TGO*** - TGO (U/mL)

##### ***Triglicerídeos*** - triglicerídeos

##### ***Plaquetas*** - plaquetas por cúbico (mL/1000)

##### ***Protrombina*** - tempo de protrombina (seg)

##### ***Estágio*** - estágio histológico da doença (1,2,3 ou 4)

## **Iniciando o Processo de Visualizações e Análises de Dados Através de Gráficos Dashboards**
### **Definindo Paletas de Cores para Gráficos mais Atrativos e Organizados**
### **Definindo "Status" como Parâmetro da Analise**
### **Visualizando Dados da Distribuição de Amostras por Status do Paciente**
### **Relembrando as Definições de Status**
### **C = Censurado**
### **CL = Censurado devido a Taxa do Fígado**
### **D = Óbito**

```python
# Definindo Paleta de Cores e o "Status" para o Gráfico
plt.style.use('dark_background')
status = data['Status'].value_counts().sort_index()
colors = ['tomato','darkorchid','teal']
```

```python
# Definindo Status de Amostras do Paciente e Visualizando
figure, axes = plt.subplots(1,2, figsize=(15,5), gridspec_kw={'width_ratios':[1.5,1]})
axes[0].barh(y=status.index, width=status.values, color=colors, edgecolor='white')
axes[0].set_xlabel('\nFrequência', fontsize='13', fontweight='bold', style='italic')
axes[0].set_ylabel('Status do Paciente', fontsize='14', fontweight='bold', style='italic')

for index, values in enumerate(status):
    axes[0].text(values+2, index, str(values), va='center', fontsize=12, fontweight='bold', style='italic')

axes[0].grid(alpha=0.4)
axes[1].pie(status.values, labels=status.index, colors=colors, explode=[0.05,0.02,0.02], autopct='%.2f%%')
figure.suptitle('Distribuição de Amostras por Status dos Pacientes\n', fontsize=15, fontweight='bold', style='italic')
plt.tight_layout()
plt.show()
```

### **O Tamanho da Amostra para o Grupo 'CL' pode ser Insuficiente para Comparar com os outros 2 grupos, pois, há poucos Registros**

```python
# Importando o Scipy Stats 
from scipy.stats import ttest_ind


# Exibindo as Estatísticas Resumidas Agrupada as Idades
display(data.groupby('Status').Age.describe())

plt.style.use('dark_background')

# Visualização
figure, axes = plt.subplots(1,2, figsize=(15,5), gridspec_kw={'width_ratios':[1.5,1]})
sns.histplot(data=data, x=data['Age'], hue=data['Status'], palette=['cyan','rebeccapurple','lawngreen'], ax=axes[0])
sns.boxplot(data=data, x=data['Status'], y=data['Age'], palette=['cyan','rebeccapurple','lawngreen'], ax=axes[1])
figure.suptitle('Distribuição da Idade do Paciente em Diferentes Status', fontsize=17, fontweight='bold', style='italic')
plt.tight_layout()
plt.show()


# Realizando Teste T Independente
# Comparando a Média de Idade entre Pacientes com Status 'Censurado' e 'Morte' (Censored - Death)
# O Tamanho da Amostra para o Grupo 'CL' pode ser Insuficiente para Comparar com os outros 2 grupos, pois, há poucos Registros
fil_df = data.groupby('Status')
censored = fil_df.get_group('C')
death = fil_df.get_group('D')

res = ttest_ind(censored['Age'],death['Age'], random_state=122)

test_stat = round(res[0], 4)
pvalue = round(res[1], 4)

print('Teste Estatístico: ', test_stat)
print('Teste P-Value: ', pvalue)

alpha=0.05
if pvalue < alpha:
    print('Há uma Diferença Estatística entre a Idade dos Pacientes Censurados e a Situação de Óbito.')
else:
    print('Evidência Insuficiente para Rejeitar a Hipótese Nula. Não há Diferença Estatisticamente Significativa entre a Idade dos Pacientes na Fase Censurada e na Fase de Óbito.')
```
